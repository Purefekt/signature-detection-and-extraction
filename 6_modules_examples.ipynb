# Example of 6 Modules

This jupyter notebook demos how to use the 6 different modules. 

All of these modules will try to extract the bounding box for the signature in a document.

Created by --> Veer Singh at Curl

```python
import cv2
from all_modules import AllModules
import IPython.display
import PIL.Image
```

## Test Image

Input parameter: Input image path

Output: Bounding box coordinate list -> [xmin, ymin, xmax, ymax]
- (xmin, ymin) -> top left edge of bbox
- (xmax, ymax) -> bottom right edge of bbox

```python
# Input
input_image_path = 'assets/test_image.tif'

img = cv2.imread(input_image_path)

# Show the image
IPython.display.display(PIL.Image.fromarray(img))
```

## Using all 6 modules

```python
module_1_bbox = AllModules(input_image_path=input_image_path).module_1()
module_2_bbox = AllModules(input_image_path=input_image_path).module_2()
module_3_bbox = AllModules(input_image_path=input_image_path).module_3()
module_4_bbox = AllModules(input_image_path=input_image_path).module_4()
module_5_bbox = AllModules(input_image_path=input_image_path).module_5()
module_6_bbox = AllModules(input_image_path=input_image_path).module_6()

print(f'Bbox values for module 1 --> {module_1_bbox}')
print(f'Bbox values for module 2 --> {module_2_bbox}')
print(f'Bbox values for module 3 --> {module_3_bbox}')
print(f'Bbox values for module 4 --> {module_4_bbox}')
print(f'Bbox values for module 5 --> {module_5_bbox}')
print(f'Bbox values for module 6 --> {module_6_bbox}')
```

## Drawing the bounding boxes

We can draw the bounding boxes using the given bbox coordinates

```python
# Draw bbox for all 6 modules in different colours
cv2.rectangle(img, (module_1_bbox[0], module_1_bbox[1]), (module_1_bbox[2], module_1_bbox[3]), (255, 0, 0), 5)
cv2.rectangle(img, (module_2_bbox[0], module_2_bbox[1]), (module_2_bbox[2], module_2_bbox[3]), (0, 255, 0), 5)
cv2.rectangle(img, (module_3_bbox[0], module_3_bbox[1]), (module_3_bbox[2], module_3_bbox[3]), (0, 0, 255), 5)
cv2.rectangle(img, (module_4_bbox[0], module_4_bbox[1]), (module_4_bbox[2], module_4_bbox[3]), (255, 255, 0), 5)
cv2.rectangle(img, (module_5_bbox[0], module_5_bbox[1]), (module_5_bbox[2], module_5_bbox[3]), (255, 0, 255), 5)
cv2.rectangle(img, (module_6_bbox[0], module_6_bbox[1]), (module_6_bbox[2], module_6_bbox[3]), (0, 255, 255), 5)

print('Some bounding boxes are overlapping')
```

```python
# Show the image
IPython.display.display(PIL.Image.fromarray(img))
```
